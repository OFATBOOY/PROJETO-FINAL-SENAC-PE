# Código para Validar CPF's, criando uma nova planilha excel contendo o resultado.

```python
import pandas as pd

#Ler a base de dados
Plan = pd.read_excel(r"Diretorio do arquivo dentro do computador")

#Conta as linhas de CPF
Quant=Plan["CPF"].count()
CountLim=Quant
CountLim=CountLim-1

#Variáveis para validação de CPF
Soma1=float
Soma2=float
Num1=float
Num2=float

#Listas para criação do novo arquivo excel com o resultado da validação
Situ=[]
Situ_inverso=[]

#Limpa os caracteres ponto e traço do CPF
while (Quant>=1):
    Limp=Plan["CPF"][CountLim]
    char=".-"
    for y in range(len(char)):
        Limp=Limp.replace(char[y],"")    
    i=1
    
    #Faz o calculo de validação do CPF
    while (i>=1):
        list[int(Limp)]
        Soma1=0
        for Mult1 in range(8):
            Posi=(int(Limp[Mult1]))
            Mult1 = Mult1+1
            Num1=Mult1*Posi
            Soma1=Soma1+Num1
        Soma1=Soma1+(9*(int(Limp[8])))    
        Mult1=0
        Mult3=9
        Soma2=0
        
        for Mult2 in range(8):
            Posi=(int(Limp[Mult2]))
            Mult2 = Mult2+1
            Num2=Mult3*Posi
            Soma2=Soma2+Num2
            Mult3=Mult3-1
        Mult2=0
        Soma2=Soma2+(1*(int(Limp[8])))
        
        Soma1=Soma1%11
        Soma2=Soma2%11
        
        #Condições para Validar CPF's como VÁLIDO OU INVÁLIDO
        
        #CPF's com números iguais
        if(Limp[0]==Limp[1]) and (Limp[1]==Limp[2]) and (Limp[2]==Limp[3]) and (Limp[3]==Limp[4]) and (Limp[4]==Limp[5]) and (Limp[5]==Limp[6]) and (Limp[6]==Limp[7]) and (Limp[7]==Limp[8]) and (Limp[8]==Limp[9]) and (Limp[9]==Limp[10]):
            Situ.append('CPF INVÁLIDO')
            
        #Válida se os CPF's são possiveis.
        elif(Soma1==(int(Limp[9]))) and (Soma2==(int(Limp[10]))):
            Situ.append('CPF VÁLIDO')
        elif(Soma1==10) and (Soma2==(int(Limp[10]))):
            Situ.append('CPF VÁLIDO')
        elif(Soma1==(int(Limp[9]))) and (Soma2==10):
            Situ.append('CPF VÁLIDO')
        elif(Soma1==10) and (Soma2==10):
            Situ.append('CPF VÁLIDO')
        #Caso não será Inválido.    
        else:
            Situ.append('CPF INVÁLIDO')
                            
        i=i-1
    
    i=i+1
    Quant=Quant-1
    CountLim=CountLim-1
    
#Leva o resultado das condições para um arquivo em Arquivo Excel.

for i in range(1,len(Situ)+1):
    Situ_inverso.append(Situ[-i])

Situa=pd.DataFrame({'CPF' : Plan['CPF'],
                    'Situação' : Situ_inverso})
print(Situa)
Situa.to_excel('ValidandoCpfAutomaticamente-SENAC.xlsx', index=0)
```
